```python
# Function For Analysing with CHI2
def chi2_sq_test(var1,var2):
  """Check for Catergorical Variable Relationship"""
  contingency_table=pd.crosstab(var1,var2)
  _,pvalue,_,_=chi2_contingency(contingency_table)
  if pvalue < 0.05 :
    print("Accept the Alternative Hypothesis", round(pvalue,2))
  else:
    print("Do Not Accept the Alternative Hypothesis", round(pvalue,2))
  return contingency_table
```



*   Null Hypothesis : Gender and Loan Status are Independant
*   Alternate Hypothesis : Gender and Loan Status are Dependant

# **Data Preparation**

```python
train.info()
```

```python
# Function for preparing Data for The Model
def data_prep(data):

  #Imputations of Missing Values
  data['Gender'] = np.where(data['Gender'].isna(), data['Gender'].mode(dropna= True)[0], data['Gender'])
  data['Married'] = np.where(data['Married'].isna(), data['Married'].mode(dropna= True)[0], data['Married'])
  data['Dependents'] = np.where(data['Dependents'].isna(), data['Dependents'].mode(dropna =True)[0], data['Dependents'])
  data['Self_Employed'] = np.where(data['Self_Employed'].isna(), data['Self_Employed'].mode(dropna= True)[0], data['Self_Employed'])
  data['LoanAmount'] = np.where(data['LoanAmount'].isna(), data['LoanAmount'].mode(dropna= True)[0], data['LoanAmount'])
  data['Loan_Amount_Term'] = np.where(data['Loan_Amount_Term'].isna(), data['Loan_Amount_Term'].mode(dropna= True)[0], data['Loan_Amount_Term'])
  data['Credit_History'] = np.where(data['Credit_History'].isna(), data['Credit_History'].mode(dropna= True)[0], data['Credit_History'])
  #Replacing the Inconsistent values
  data['Dependents'].replace('3+', 3, inplace= True)

  #Converting the DataType
  data['Credit_History'] = data['Credit_History'].astype('int')
  data['Dependents'] = data['Dependents'].astype('int')

  return data
```

# **Encoding**

```python
# Function For Encoding Data
def encode(data):
  le = LabelEncoder()
  for i in data.columns:
    if data[i].dtype == 'object':
      data[i] = le.fit_transform(data[i])
  return data
```

# **Splitting Of Data**

# Decision Tree Classifier

```python
dtree = DecisionTreeClassifier()

#Fit the Model
dtree.fit(x_train, y_train)
```

```python
y_hat_train_dtree = dtree.predict(x_train)
y_hat_test_dtree = dtree.predict(x_test)
```

```python
def model_eval(actual , predicted):
  conf_matrix = confusion_matrix(actual , predicted)
  clas_rep = classification_report(actual , predicted)
  acc_score = accuracy_score(actual , predicted)

  print("The Accuracy of the Model is:" , round(acc_score , 2))
  print(conf_matrix)
  print(clas_rep)
```

```python
model_eval(y_train , y_hat_train_dtree)
```

```python
model_eval(y_test , y_hat_test_dtree)
```
